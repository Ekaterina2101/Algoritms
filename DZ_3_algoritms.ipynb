Задание 2 Подберите аргументы функции eval_model для логистической регрессии таким образом, чтобы log loss был минимальным

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([ [   1,    1,  500,    1],
               [   1,    1,  700,    1],
               [   1,    2,  750,    2],
               [   1,    5,  600,    1],
               [   1,    3, 1450,    2],
               [   1,    0,  800,    1],
               [   1,    5, 1500,    3],
               [   1,   10, 2000,    3],
               [   1,    1,  450,    1],
               [   1,    2, 1000,    2]], dtype=np.float64)

y = np.array([0, 0, 1, 0, 1, 0, 1, 0, 1, 1], dtype=np.float64)
```

```python
def standard_scale(x):
    res = (x - x.mean()) / x.std()
    return res

X_st = X.copy()
X_st[:, 2] = standard_scale(X[:, 2])
```

```python
def calc_logloss(y, y_pred):
    err = - np.mean(y * np.log(y_pred) + (1.0 - y) * np.log(1.0 - y_pred))
    return err
```

```python
def sigmoid(z):
    res = 1 / (1 + np.exp(-z))
    return res
```

```python
def eval_model(X, y, iterations, eta=1e-4):
    np.random.seed(42)
    W = np.random.randn(X.shape[1])
    n = X.shape[0]
    
    for i in range(iterations):
        z = np.dot(X, W)
        y_pred = sigmoid(z)
        err = calc_logloss(y, y_pred)
        
        
        dQ = 1/n * X.T @ (y_pred - y)
        W -= eta * dQ
        if i % (iterations / 10) == 0:
            print(i, W, err)
    return W
```

```python
W = eval_model(X_st, y, iterations=500, eta=1)
W.shape
```

```python
W = eval_model(X_st, y, iterations=35000, eta=1)
```

```python
W = eval_model(X_st, y, iterations=1000, eta=10)
```
